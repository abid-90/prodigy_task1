```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("dataset1.csv")
```

```python
data.head()
```

```python
data.describe
```

```python

countries = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

filtered_data = data[data['Country Name'].isin(countries)]

plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.show()

```
